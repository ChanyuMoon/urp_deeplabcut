## Preprocessing the collected data

```python
import pandas as pd
import os
# 정규식을 통한 search
import re
```

### 0. Loading data

```python
dir_path = os.getcwd()
print(dir_path)
df_1 = pd.read_csv(dir_path + '/data/CollectedData_CYMoon_trial_1.csv', header=1, index_col=0)
df_2 = pd.read_csv(dir_path + '/data/CollectedData_CYMoon_trial_2.csv', header=1, index_col=0)
```

```python
df_1.columns
```

```python
df_1.index
```

```python
df_2.index
```

```python
def transform_index(raw_index_iter, pattern, group_num_to_extract=0):
    index_list = []
    for index in raw_index_iter:
        try:
            frame_num = pattern.search(index)[group_num_to_extract]
#             print(frame_num)
            index_list.append(frame_num)
        except:
            print(f'No match in {index}')
    return index_list

p = re.compile('(img)([0-9]{3})(.[a-zA-Z]+)')
index_list_1 = transform_index(df_1.index, p, 2)
index_list_2 = transform_index(df_2.index, p, 2)
```

```python
print(index_list_1)
print(len(index_list_1))
print(index_list_2)
print(len(index_list_2))
```

```python
df_1 = df_1.drop('coords')
```

```python
df_1.head()
```

```python
print(df_1.shape)
```

### 1. Making multi column

```python
p = re.compile('[a-zA-Z]+_?[0-9]?')
column_1 = transform_index(df_1.columns, p)
column_1 = [column_1, ['x','y']*26]
print(len(column_1[0]))
print(len(column_1[1]))
print(column_1)
column_2 = transform_index(df_2.columns, p)
column_2 = [column_2, ['x','y']*26]
print(len(column_2[0]))
print(len(column_2[1]))
print(column_2)
print(column_1 == column_2)

columns = column_1.copy()
print(columns)
```

### 2. Column, index change
